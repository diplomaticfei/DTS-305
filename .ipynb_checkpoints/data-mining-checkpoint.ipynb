```python
# Import librares
import requests  # To get data from website
from bs4 import BeautifulSoup  # To read HTML
import pandas as pd  # To work with data tables
import matplotlib.pyplot as plt  # To make charts
```

```python
# The website we want to scrape
url = "http://books.toscrape.com/"

# Try to get the webpage
try:
    response = requests.get(url)
    print("‚úÖ Successfully connected to website!")
    print(f"Status Code: {response.status_code}")
except:
    print("‚ùå Could not connect to website")
    print("Check your internet connection")
```

```python
# Let's see what the website looks like in HTML
soup = BeautifulSoup(response.content, 'html.parser')

# Find all book containers (each book is in an 'article' tag)
books = soup.find_all('article', class_='product_pod')
print(f"Found {len(books)} books on the page!")
```

```python
# Let's practice on just one book first
first_book = books[0]

# Get the title
title = first_book.find('h3').find('a')['title']
print(f"Title: {title}")

# Get the price
price = first_book.find('p', class_='price_color').text
print(f"Price: {price}")

# Get the rating
rating_tag = first_book.find('p', class_='star-rating')
rating_class = rating_tag['class'][1]  # Gets 'Three', 'Five', etc.
print(f"Rating: {rating_class}")

# Get availability
availability = first_book.find('p', class_='instock availability').text.strip()
print(f"Availability: {availability}")
```

```python
# List to store all book data
all_books = []

# Loop through each book
for book in books:
    try:
        # Get title
        title_tag = book.find('h3').find('a')
        title = title_tag['title']
        
        # Get price
        price_tag = book.find('p', class_='price_color')
        price_text = price_tag.text
        
        # Get rating
        rating_tag = book.find('p', class_='star-rating')
        rating_class = rating_tag['class'][1]
        
        # Get availability
        availability_tag = book.find('p', class_='instock availability')
        availability = availability_tag.text.strip()
        
        # Add to our list
        all_books.append({
            'Title': title,
            'Price': price_text,
            'Rating': rating_class,
            'Availability': availability
        })
    except:
        # Skip if there's an error with one book
        continue

print(f"‚úÖ Successfully scraped {len(all_books)} books!")
```

```python
# Create a table from our data
df = pd.DataFrame(all_books)

# Show the first 5 books
print("First 5 books in our table:")
print(df.head())

# Show basic info
print("\nüìä Our Data Table Info:")
print(f"We have {len(df)} books")
print("Columns:", list(df.columns))
```

```python
# Clean the price column
def clean_price(price_str):
    # Remove ¬£ symbol
    clean = price_str.replace('¬£', '')
    # Convert to float (decimal number)
    return float(clean)

# Apply cleaning to all prices
df['Price_Number'] = df['Price'].apply(clean_price)

print("‚úÖ Prices cleaned!")
print("First few cleaned prices:")
print(df[['Price', 'Price_Number']].head())
```

```python
# Create a dictionary to convert words to numbers
rating_dict = {
    'One': 1,
    'Two': 2,
    'Three': 3,
    'Four': 4,
    'Five': 5,
    'Zero': 0
}

# Convert ratings to numbers
df['Rating_Number'] = df['Rating'].map(rating_dict)

print("‚úÖ Ratings converted to numbers!")
print("Rating conversion examples:")
print(df[['Rating', 'Rating_Number']].head(10))
```

```python
# Create a new column for price category
df['Price_Category'] = 'Expensive'  # Start with all expensive

# Change to Affordable if price < ¬£20
df.loc[df['Price_Number'] < 20, 'Price_Category'] = 'Affordable'

print("‚úÖ Added price categories!")
print("\nHow many in each category?")
print(df['Price_Category'].value_counts())
```

```python
# Calculate average rating for affordable books
affordable_avg = df[df['Price_Category'] == 'Affordable']['Rating_Number'].mean()

# Calculate average rating for expensive books
expensive_avg = df[df['Price_Category'] == 'Expensive']['Rating_Number'].mean()

print("üìä Average Ratings:")
print(f"Affordable books (< ¬£20): {affordable_avg:.2f} stars")
print(f"Expensive books (‚â• ¬£20): {expensive_avg:.2f} stars")
```

```python
# Count books in each category
counts = df['Price_Category'].value_counts()

# Create pie chart
plt.figure(figsize=(8, 6))
plt.pie(counts.values, labels=counts.index, autopct='%1.1f%%', 
        colors=['lightgreen', 'lightcoral'], startangle=90)
plt.title('Affordable vs Expensive Books', fontsize=16, fontweight='bold')
plt.axis('equal')  # Makes the pie circular
plt.tight_layout()
plt.show()

print("‚úÖ Pie chart created!")
```

```python
# Data for the bar chart
categories = ['Affordable', 'Expensive']
averages = [affordable_avg, expensive_avg]
colors = ['lightgreen', 'lightcoral']

# Create bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(categories, averages, color=colors, edgecolor='black')

# Add value labels on top of bars
for bar, value in zip(bars, averages):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.05,
             f'{value:.2f}', ha='center', va='bottom', fontsize=12)

# Add chart details
plt.title('Average Ratings by Price Category', fontsize=16, fontweight='bold')
plt.xlabel('Price Category', fontsize=12)
plt.ylabel('Average Rating (1-5 stars)', fontsize=12)
plt.ylim(0, 5.5)  # Set y-axis limit (ratings are 1-5)
plt.grid(axis='y', alpha=0.3)  # Light grid lines
plt.tight_layout()
plt.show()

print("‚úÖ Bar chart created!")
```

```python
# Save to CSV file
df.to_csv('books_data_final.csv', index=False)
print("‚úÖ Data saved as 'books_data_final.csv'")

# Show what we saved
print("\nüìÅ Files created:")
print("1. books_data_final.csv - All the book data")
print("2. Your Jupyter Notebook with all the code")
```
